# Clases de Machine Learning

```python
import os, sys
import pandas as pd 
import numpy as np 
import seaborn as sns
import pandas_gbq
import matplotlib.pyplot as plt 
```

```python
#Definir las rutas de los directorios 
from pathlib import Path
current_dir = Path('.').resolve()
project_dir = Path(current_dir).parent 
```

```python
query = """
SELECT  
  *
FROM 
  `bigquery-public-data.chicago_taxi_trips.taxi_trips` 

LIMIT 
  1000
"""
df = pd.read_gbq(query)
df.head()
```

```python
# get only numeric colums 
numerics = ['int16','int32','int64','float16','float32','float64']
num_df = df.select_dtypes(include=numerics)
num_df.head()
```

```python
num_df.describe()
```

```python
cols = ['trip_total', 'trip_seconds', 'tips', 'fare', 'extras']
num_df[cols]
```

```python
fig, ax = plt.subplots(ncols=5, nrows=1, figsize=(30, 5))

log_data = pd.DataFrame(columns=cols)
for i, col in enumerate(num_df[cols] ):
    np.log(num_df[col] + 1).hist(ax=ax[i])
    ax[i].set_title(col)
    
    log_data[col] = np.log(num_df[col] + 1)
```

# Transform Data

```python
X = log_data[cols].values #trasformamos de pandas a numpy 
```

```python
type(X)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X)
X_std = scaler.transform(X)
```

```python
X_std.shape #1000 filas y 5 columnas 
```

# Matriz de covarianza 

```python
cov_mat = np.cov(X_std) #Matriz de covarianza 
```

```python
cov_mat.shape
```

# Obtener los autovalores y autovectores "eigh valores"

```python
eigen_vals, eigen_vecs = np.linalg.eigh(cov_mat) #valores propios y vectores propios 
```

```python
print('\n Autovalores \n {}'.format(eigen_vals))
```

Selecciono los datos mas grandes en este caso los ultimos 3, quire decir que paso puedo pasar de 5 dimensiones a 3

```python
eigen_vals_sort = eigen_vals[::-1] #ordenar los datos de mayor a menor 
```

```python
plt.plot(eigen_vals_sort[:10])
```

```python
total = np.sum(eigen_vals_sort)
total 
```

```python
#Cuanta varianza de cada auto valor, entonces se divide cada valor por la suma
var_exp = [(i/total) for i in eigen_vals_sort]
```

```python
plt.plot(var_exp[:10]) #Aca puedo ver que porcentaje ocupa cada dato, la primera explica el 40% de la varianza la segunda como el 35%,
# con la tercera el 20%,es decir que con esas 3 variables explico el 95%
```

```python
plt.plot(np.cumsum(var_exp[:10])) # aca grafico la suma acumulativa de los eigen valores
```

```python
x=np.arange(1, 11)
y=var_exp[:10]

plt.bar(x, y, label='Varianza explicada individual')
plt.step(x, np.cumsum(y),color='red', where='mid',label='Varianza explicada acumulada')

plt.ylabel('Explained varianza ratio')
plt.xlabel('Principal Component index')

plt.legend()
```

```python
#vamos a pasar de dimesion 5 a dimension a 3
```

# Calcular PCA

```python
from sklearn.decomposition import PCA 
pca = PCA(n_components = 3).fit(X_std) #El tres viene del analisis anterior de eigen valores 
X_pca = pca.transform(X_std)
```

```python
pca.explained_variance_ratio_.sum()
```
